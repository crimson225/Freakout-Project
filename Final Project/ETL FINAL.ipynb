```python
dowdf2 = dowdf[['Date','Open','High','Low','Close','Volume']].copy()
dowdf2 = dowdf2.rename(columns={'Date':'thisday','Open':'dow_open','High':'dow_high','Low':'dow_low','Close':'dow_close','Volume':'dow_volume'})
dowdf2.head()
```

```python
bud = "../csv_files/BUD.csv"
buddf = pd.read_csv(bud)
buddf.head()
```

```python
newdf = buddf[['Date','Open','High','Low','Close','Volume']].copy()
newdf = newdf.rename(columns={'Date':'thisday','Open':'bud_open','High':'bud_high','Low':'bud_low','Close':'bud_close','Volume':'bud_volume'})
newdf.head()
```
