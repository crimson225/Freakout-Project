```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('2019_nCoV_data.csv')
df
```

```python
df['Date'] = pd.to_datetime(df['Date']).dt.date
df
```

```python
df.columns
```

```python
df_1 = df.groupby(df['Date'])['Confirmed'].sum()
df_1
```

```python
df_1.to_csv('confirmed_cases.csv')
```

```python
df_1.plot(kind='line')
plt.show
```
